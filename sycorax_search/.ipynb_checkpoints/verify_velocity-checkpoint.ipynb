```python
import numpy as np
import pylab as plt
import pandas as pd
```

```python
## load in the data, created using prep_csv.py
infile_name = '/home/ellie/research/lsst/s1003Hmna_data_vel.csv'
df = pd.read_csv(infile_name)
```

```python
## plot the velocities overlaid on each other

## first, plot the Keplerian and "measured"
## velocity magnitudes for comparison

plt.scatter(df['r'], df['vk'], marker='.', label='kepler')
plt.scatter(df['r'], df['v'], marker='.', label='measured')
plt.title('Velocity Magnitude Comparison')
plt.legend()

plt.xlabel('Heliocentric distance (AU)')
plt.ylabel('Velocity (AU/day)')
plt.show()
```

```python
plt.scatter(df['r'], df['x_dotk'], marker='.', label='kepler')
plt.scatter(df['r'], df['heliocentricVX'], marker='.', label='measured')
plt.title('X Velocity Comparison')
plt.legend()

plt.xlabel('Heliocentric distance (AU)')
plt.ylabel('Velocity (AU/day)')
plt.show()
```

```python
## plot the velocities overlaid on each other

plt.scatter(df['r'], df['y_dotk'], marker='.', label='kepler')
plt.scatter(df['r'], df['heliocentricVY'], marker='.', label='measured')
plt.title('Y Velocity Comparison')
plt.legend()

plt.xlabel('Heliocentric distance (AU)')
plt.ylabel('Velocity (AU/day)')
plt.show()
```

```python
## plot the velocities overlaid on each other

plt.scatter(df['r'], df['z_dotk'], marker='.', label='kepler')
plt.scatter(df['r'], df['heliocentricVZ'], marker='.', label='measured')
plt.title('Z Velocity Comparison')
plt.legend()

plt.xlabel('Heliocentric distance (AU)')
plt.ylabel('Velocity (AU/day)')
plt.show()
```
