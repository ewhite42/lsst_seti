```python
## use this program to estimate maximum accelerations of objects
## assuming a range of values for thrust
```

```python
import numpy as np
import pylab as plt
```

```python
## define a range of thrust values

## comets exhibit "exhaust velocities" on the order of 0.1 to 1.0 km/s
## rockets exhibit exhaust velocities of 2.5-4.5 km/s

## comets exhibit mass loss rates of 1-1000 kg/s when active
## rockets exhibit mass loss rates of 270-2800 kg/s

## rockets experience typical thrusts of 2-8 MN
## comets experience typical thrusts of about 10^-5 to 10^-1 N

## Thrusts are given here in Newtons
thrusts = np.logspace(-5, 7, num=13)
```

```python
print(thrusts)
```

```python
## determine a starting mass
## comet masses go up to 10^17 kg
## rocket masses go up to a few million kg
## let's use comets as the upper limit

masses = np.logspace(-17, 17, num=35)
print(masses)
```

```python
## gravitational constant of the Sun, in m^3/s^2
G = 1.3271244*(10**20)

nga_sets = []

for i in range(len(thrusts)):
    ngas = abs((thrusts[i]/masses - G)*0.0499001755) ## convert to au/day^2
    nga_sets.append(ngas)
    #print(ngas)
    plt.loglog(masses, ngas, label=str(np.log10(thrusts[i])))
    
plt.legend()
plt.show()
```
