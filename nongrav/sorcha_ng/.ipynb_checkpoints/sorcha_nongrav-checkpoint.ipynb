```python
## In this notebook, we will walk through steps of implementing 
## nongravitational accelerations in Sorcha, plotting the results
## after various steps for verification

## MEW, 12 October 2025

import pylab as plt
import pandas as pd
```

```python
## read in the csv file resulting from running Sorcha on the
## C2017 M4 ATLAS data with only gravitational acceleration

df = pd.read_csv('/home/ellie/research/lsst/sorcha_output/c2017m4_atlas.csv')

x = df['fieldMJD_TAI'] 
y1 = df['RA_deg']
y2 = df['Dec_deg']

plt.plot(x,y1)
plt.xlabel('Field MJD')
plt.ylabel('RA (deg)')
plt.show()

plt.plot(x,y2)
plt.xlabel('Field MJD')
plt.ylabel('RA (deg)')
plt.show()
```
